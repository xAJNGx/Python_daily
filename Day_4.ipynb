```python
def fibo(n):
    a,b = 0,1
    result = []
    for i in range(n+1):
        result.append(b)
        a,b = b, a+b
    return result

fibo(5)
```

```python
a = "eve"

if a == a[::-1]:
    print("palindrome")
else:
    print("Not a palindrome")
```

```python
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"
```

```python
def calculate_area(length, width):
    return length * width

print(calculate_area(10, 5))  

# Using keyword arguments
print(calculate_area(width=5, length=10))
```

```python
def sum_numbers(*args):
    return sum(args)

print(sum_numbers(1, 2, 3, 4))  
print(sum_numbers(5, 10))
```

```python
def display_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

display_info(name="Anuj", age=22, country="Nepal")
```

```python
def mixed_function(arg1, arg2="default", *args, **kwargs):
    print(f"arg1: {arg1}, arg2: {arg2}")
    print(f"args: {args}")
    print(f"kwargs: {kwargs}")

mixed_function(10, 20, 30, 40, name="Anuj", age=22)
```
